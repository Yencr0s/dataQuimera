```python
os.chdir('/content/drive/MyDrive/data')
with open('val.json') as f:
  info=json.loads(f.read())
# dict_keys(['info', 'license', 'images', 'annotations', 'categories'])
```

```python
animales=['aguila','ave','felinos','lagarto','perro','polar','roedor']
for animal in animales:
  print(animal)
  os.chdir('/content/drive/MyDrive/data/'+animal)
  res=dict()
  res['info']=info['info']
  res['license']=info['license']
  res['categories']=info['categories']
  arr=[]
  ids=[]
  trad=dict()
  for elemx in info['images']:
    elem=elemx.copy()
    if os.path.isfile(elem['file_name']):
      trad[elem['id']]=len(ids)
      ids.append(elem['id'])
      elem['id']=trad[elem['id']]
      arr.append(elem)
  print('num_fotos',len(arr))
  res['images']=arr
  arr=[]
  for annx in info['annotations']:
    ann=annx.copy()
    if ann['image_id'] in ids:
      ann['image_id']=trad[ann['image_id']]
      ann['id']=len(arr)
      arr.append(ann)
  print('num_partes',len(arr))
  res['annotations']=arr
  with open('anot.json','w') as f:
    json.dump(res,f,indent=4)
```
