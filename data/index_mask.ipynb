```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
cabezas=[0,4,13,20] # 4 patas, 2 patas, pajaros, reptiles
cuerpos=[21,14,15,5,1]
patas=[22,16,6,7,2]
colas=[23,17,3]

animales=['aguila','ave','felinos','lagarto','perro','polar','roedor']
for animal in animales:
  os.chdir('/content/drive/MyDrive/data/'+animal)
  with open('anot.json') as f:
    info=json.loads(f.read())
  res=dict()
  res['n_total']=len(info['images'])
  res['animales']=[]
  for i in range(len(info['images'])):
    elem=dict()
    cabeza=[]
    cuerpo=[]
    pata=[]
    cola=[]
    index=0
    for j in range(len(info['annotations'])):
      aux=info['annotations'][j]
      if aux['image_id']!=i:
        continue 
      if aux['category_id'] in cabezas:
        cabeza.append(index)
      elif aux['category_id'] in cuerpos:
        cuerpo.append(index)
      elif aux['category_id'] in patas:
        pata.append(index)
      elif aux['category_id'] in colas:
        cola.append(index)
      else:
        print(aux['category_id'])   
      index+=1
    elem['cabeza']=cabeza
    elem['cuerpo']=cuerpo
    elem['patas']=pata
    elem['cola']=cola
    elem['file_name']=info['images'][i]['file_name']
    res['animales'].append(elem)

    with open('res.json','w') as f:
      json.dump(res,f,indent=4)
```
