```python
from pycocotools.coco import COCO
import os
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import json
animales=['aguila','ave','felinos','lagarto','perro','polar','roedor']
partes=['cuerpo','patas']
for animal in animales:
  print('> ',animal)
  os.chdir('/content/drive/MyDrive/data/'+animal)
  with open('res.json') as f:
    info=json.loads(f.read())
  coco = COCO('anot.json')
  img_dir = ''

  n = info['n_total']
  for image_id in range(n):
    elem=info['animales'][image_id]
    img = coco.imgs[image_id]
    image = np.array(Image.open(os.path.join(img_dir, img['file_name'])))

    cat_ids = coco.getCatIds()
    anns_ids = coco.getAnnIds(imgIds=img['id'], catIds=cat_ids, iscrowd=None)
    anns = coco.loadAnns(anns_ids)

    for parte in partes:
      mask = coco.annToMask(anns[0])>777
      for i in elem[parte]:
        mask += coco.annToMask(anns[i])>0
      img=Image.fromarray(mask)
      img.save(parte+'/'+elem['file_name']) 
```
